### To find k-th largest element in the array with length n, quick sort algo efficiency is only n*log(n) time complexity

```python
def quick_sort(arr):
    if len(arr) <= 1:
        return arr
    else:
        middle = arr[0]
        less = [x for x in arr[1:] if x <= middle]
        more = [x for x in arr[1:] if x > middle]
        return quick_sort(less) + [middle] + quick_sort(more)
    
 
arr = [3, 1, 5, 12, 2, 11, 4, 8, 10, 7]
k = 3
sorted_arr = quick_sort(arr)
print(f"{k}-th largest element:", sorted_arr[-k])
```

### via min heap with only k largest elements, finding k-th largest elements in a array has n*log(k) time complexity

It is minimum heap. In the tree top we have the smallest element. Element with the index i is always smaller than 2i+1-th and 2i+2-th elements.

If there are 2i+1 elements in a heap, the newly added element index will be 2i+2 in the begginng and the parent of the new element is i-th element. Then we compare i-th element and 2i+2-th element. If 2i+2-th element is smaller, we swap them and we will do it until the whole heap will become minimum heap again.

When there 2i+1 elements in a heap and are we delete the element with index j, we put 2i+1-th element in the index j. We swap elements, until the heap becomes minimum heap again.

**Solution**. We keep only k elements in the heap and we try these elements to be the largest among all elements in a array. We iterate through all elements in a array, if new element is larger than tree top element of the heap, we throw tree top element out and add new element to the heap. At the end, the k-th largest element of the array is the tree top element.

```python
import heapq  # it's a minimum heap library


def find_k_largest(nums, k):
    heap = nums[:k]
    # we try to keep the size of the heap equal to k
    heapq.heapify(heap)

    for num in nums[k:]:
        if num > heap[0]:  # larger than smallest element of the heap
            # we remove the smallest element from the heap
            # to have opportunity to add new element to the heap
            heapq.heappop(heap)
            # Insert elements into the priority queue
            heapq.heappush(heap, num)        
    
    return heap[0]


# Given array
arr = [3, 1, 5, 12, 2, 11, 4, 8, 10, 7]
# Size of array
k = 3

print(f"{k}-th largest element:", find_k_largest(arr, k))
```
